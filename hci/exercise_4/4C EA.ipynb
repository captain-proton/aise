```python
# Task 4C.3:
# Implement a mutation of vector1
# The function should take `vector1` as an argument and return a new ColorBinVector that has 
# (for example) two of its bits flipped at a random position within the vector.
# Hint: use random.randint(0, len(vector1)) twice but check for inequal results
# Hint: Use get_bin() and set_bin() of ColorBinVector

def mutation(vector1: ColorBinVector, flip=2) -> ColorBinVector:
    binary = list(vector1.get_bin())
    #print(''.join(binary))
    
    flip_indexes = random.sample(range(len(binary)), flip)
    #print(flip_indexes)
    for index in flip_indexes:
        binary[index] = "0" if binary[index] == "1" else "1"
    vector1.set_bin(''.join(binary))
    return vector1

v = ColorBinVector()
print(mutation(v).get_bin())
```

```python
# Task 4C.4:
# Implement a function that creates a crossover between two vectors
# This function should combine some part of vector1 with some other part of vector2.

def crossover(vector1: ColorBinVector, vector2: ColorBinVector) -> ColorBinVector:
    b1 = vector1.get_bin()
    b2 = vector2.get_bin()
    
    binary = ''.join([b1[0:8], b2[8:16], b1[16:24]])
    
    v = ColorBinVector()
    v.set_bin(binary)
    return v

v = ColorBinVector()
w = ColorBinVector()
print(v.get_bin())
print(crossover(v, w).get_bin())
```

```python
class SurvivalOfTheNFittest():
    
    def __init__(self, beeings: list,
                 mutation_chance=50,
                 flip_count=lambda: 2):
        super().__init__()
        self.beeings = beeings
        self.mutation_chance = 50
        self.flip_count = flip_count
        
    def mutate(self):
        
        for i in range(len(self.beeings)):
            
            if random.randint(1, 100) <= self.mutation_chance:
                
                beeing = mutation(self.beeings[i], self.flip_count())
                self.beeings[i] = beeing
    
    def evolve(self) -> list:
        
        children = []
        children += [crossover(self.beeings[0], beeing) for beeing in self.beeings[1:6]]
        children += [crossover(self.beeings[1], beeing) for beeing in self.beeings[2:7]]
        self.beeings = children
        
        if self.mutation_chance > 0:
            self.mutate()
        
        return self.beeings
        
```

```python
class Centipede(SurvivalOfTheNFittest):
    
    def __init__(self, beeings: list,
                 mutation_chance=50,
                 flip_count=lambda: 2):
        super().__init__(beeings, mutation_chance, flip_count)
        
    def evolve(self) -> list:
                
        # prepend the second beeing to produce a sibling
        self.beeings.insert(0, self.beeings[1])
        
        # pair each beeing with the next inside the population
        self.beeings = [crossover(self.beeings[i], self.beeings[i + 1]) 
                        for i in range(len(self.beeings) - 1)]
        
        if self.mutation_chance > 0:
            self.mutate()
        
        return self.beeings
```
